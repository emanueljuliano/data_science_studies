## The Pandas Series Objects

One-dimensional array of indexed data

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
```

```python
data.values
```

```python
data.index
```

```python
data[1]
```

```python
data[1:3]
```

### Series as generalized NumPy array

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=['a', 'b', 'c', 'd'])
data
```

```python
data['b']
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0], index=[2, 5, 3, 7])
```

```python
data
```

```python
data[5]
```

### Series as specialized dictionary

```python
population_dict= {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict)
population
```

```python
population['California']
```

```python
population['California':'Illinois']
```

Series seems like dictionarys but has the slicing operation and other stuffs


### Constructing Series objects

```python
pd.Series([2, 4, 6])
```

Data can be a scalar, which is repeated to fill the specified index:

```python
pd.Series(5, index=[100, 200, 300])
```

data can be a dictionary, in which index defaults to the sorted dictionay keys

```python
pd.Series({2:'a', 1:'b', 3:'c'})
```

```python
pd.Series({2:'a', 1:'b', 3:'c'}, index=[3, 2])
```

## The Pandas DataFrame Object

### DataFrame as a generalized NumPy array

DataFrame is analog of a two-dimensional array with both flaxible row indices and flexivle collumn names

```python
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict)
area
```

```python
states = pd.DataFrame({'population': population, 'area': area})
states
```
